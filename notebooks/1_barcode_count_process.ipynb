```python
def csv_read(exp):
    df_count = pd.read_csv(f"{count_raw}\\{exp}_main_barcodes_counts_unfiltered.tsv", sep='\t').rename(
        columns = {"Unnamed: 0":"label"})
    df_count.fillna(0, inplace=True)
    df_one = df_count.iloc[:, 1:].astype(int)
    df_two = pd.concat([df_count["label"], df_one], axis = 1)
    df_two = df_two.loc[df_two["c_0"] >= 10].reset_index(drop=True)     #default filter: >=10 counts in T0.
    return df_two
```

```python
def count_summary(df_count):
    summary = pd.DataFrame(df_count.iloc[:, 1:].sum(), columns=['count']).astype(int)
    summary.reset_index(inplace=True)
    summary.rename(columns={'index': 'tp'}, inplace=True)
    return summary
```

```python
def count2enrich2(df_count, experiment):
    for t in range(1, len(df_count.columns)):
        df_one = df_count.iloc[:, [0, t]]
        df_two = df_one.copy()
        df_two.rename(columns={"label": "", f"{df_two.columns[1]}": "count"}, inplace = True)
        df_two.to_csv(f"{count_processed}\\{experiment}_T{str(t-1)}.tsv", sep='\t', index=False)
    return
```

```python
experiment = "lowADSL_dox+"
df_count = csv_read(experiment)
count2enrich2(df_count, experiment)
print(count_summary(df_count))
```
